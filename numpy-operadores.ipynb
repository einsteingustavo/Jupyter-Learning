````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

## Módulo linalg (Álgebra Linear)

````python
# As funções inner() e dot() calculam o produto interno
# entre 2 vetores (recomendo o uso da 1a)
a = np.array([1, 2, -1])
b = np.array([-7, 1, -5])
print("Calculando o produto interno com inner():", np.inner(a, b))
print("Calculando o produto interno com dot():  ", np.dot(a, b))
````

````python
# A função vdot() calcula o produto interno de um vetor (o segundo parâmetro)
# com o complexo conjugado de outro (o primeiro parâmetro)
a = np.array([1 + 1j, -3 + 4j, -1j])
b = np.array([-7, 1, -5])
print("Calculando o produto interno com vdot():", np.vdot(a, b))
````

````python
# Para multiplicar arrays 2-D como matrizes, nós podemos multiplicar 
# usando a função dot() ou a mais recente matmul()
A = np.array([[1, 0], [0, 1]])
B = np.array([[4, 1], [2, 2]])
print("Calculando o produto de matrizes com dot():   ", np.dot(A, B))
print("Calculando o produto de matrizes com matmul():", np.matmul(A, B))

# Observe que Python agora tem um operador, o @, que faz multiplicação de matriz
print("Calculando o produto de matrizes com @:", A @ B)
````

````python
# Podemos calcular os autovalores e autovetores de uma matriz
# quadrada facilmente
A = [[0, 1], [-2, -3]]
e, V = np.linalg.eig(A)
print("Os autovalores de A são", e)
print("A matriz de autovetores de A é", V)
print("A @ V =", A @ V)
print("V @ e =", V @ np.diag(e))
````

````python
# O módulo linalg também tem a função matrix_power(), que efetua
# a potenciação de uma matriz quadrada
A = [[0, -1], [1/2, 1]]
print("A 10ª potência de A é", np.linalg.matrix_power(A, 10))
````

````python
# Podemos também resolver um sistema de equações lineares.  Por exemplo,
# a solução do sistema dado pelas equações 3*x + y = 9 and x + 2*y = 8 é
a, b = np.array([[3, 1], [1, 2]]), np.array([9, 8])
x = np.linalg.solve(a, b)
print("Solução é: ", x)

# Checa que a solução está correta
print("a @ x == b = ", np.allclose(a @ x, b))
````

````python
# Numpy também pode calcular a inversa de uma matriz
a_inv = np.linalg.inv(a)
print("a_inv@b = ", a_inv@b)
````

## Combinação de arrays

````python
# Podemos concatenar arrays através da função concatenate.  Os arrays
# devem ter o mesmo formato, exceto no eixo que estiverem sendo
# concatenados
a = np.arange(12).reshape(3, 4)
b = np.arange(100, 108).reshape(2, 4)
print("Formato de a = ", a.shape)
print("Formato de b = ", b.shape)

c = np.concatenate((a, b), axis=0)
print("Formato de c = ", c.shape)
print("c =\n", c)
````

````python
# Outro exemplo de concatenação
a = np.arange(12).reshape(4, 3, order="f")
b = np.arange(100, 108).reshape(4, 2, order="f")
print("Formato de a = ", a.shape)
print("Formato de b = ", b.shape)

c = np.concatenate((a, b), axis=-1)
print("Formato de c = ", c.shape)
print("c =\n", c)
````

````python
# Similar a concatenação, temos o empilhamento (stack).  Neste
# caso, uma dimensão nova é criada.  Os arrays devem ter o mesmo
# formato
a = np.arange(8)
b = np.arange(100, 108)
print("Formato de a = ", a.shape)
print("Formato de b = ", b.shape)

c = np.stack((a, b), axis=1)
print("Formato de c = ", c.shape)
print("c =\n", c)
````

## Módulo FFT

Análise de Fourier é uma técnica para expressar uma função como uma soma de componentes periódicos compostos de senos e cossenos cujas frequências são múltiplas de uma frequência fundamental.  Como a correspondência é biunívoca, a partir de uma função podemos determinar sua transformada ou série de Fourier, e a partir de sua transformada ou série podemos determinar a função.  No primeiro caso usamos a transformada de Fourier e no segundo caso, a transformada inversa de Fourier.

Quando tanto a função como sua transformada de Fourier são substituídas pela suas versões discretas, temos a Transformada Discreta de Fourier (em inglês, Discrete Fourier Transform, ou DFT).  O algoritmo para calcular esta transformada e sua inversa tem uma versão otimizada que ficou conhecida como Transformada Rápida de Fourier (em inglês, Fast Fourier Transform, ou FFT). A FFT permitiu o cálculo rápido da DFT, e com isso ajudou a popularizar a Análise de Fourier e a Análise Espectral, e impulsionou o desenvolvimento do Processamento Digital de Sinais.

Tanto a biblioteca Numpy como a Scipy possuem implementações do algoritmo de FFT, através das funções ```fft()``` e ```ifft()```, que podem ser usadas para calcular a DFT e a DFT inversa de uma sequência de pontos.

Quando um sinal é real, sua DFT possui possui uma forte simetria: as componentes da DFT para frequências negativas são o complexo conjugado das componentes de frequência positiva.  Assim, computacionalmente precisamos apenas calcular metade das componentes da transformada ou de sua inversa.

Em Numpy isto é feito pelas funções `rfft()` e `rifft()` e suas funções similares.

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Cria uma função 
freqs = [1, 35, 60]
T_MAX, FFT_LEN = 2, 4096
t = np.linspace(0, T_MAX, FFT_LEN)
yt = sum(np.sin(2*np.pi*f*t) for f in freqs)
p = plt.plot(t, yt)
````

````python
ff = np.arange(FFT_LEN)
yf = np.fft.rfft(yt)
N_POINTS = 150
c = plt.bar(ff[:N_POINTS], np.absolute(yf[:N_POINTS]))
````

````python
yf_filtered = np.zeros_like(yf)
yf_filtered[2] = yf[2]
yt_filtered = np.fft.irfft(yf_filtered)
plt.plot(t, yt)
m = plt.plot(t, yt_filtered, "r", linewidth=2)
````
