# Módulo skimage

Um módulo relacionado ao `scipy.ndimage` é o `scikit-image`, ou `skimage` de forma abreviada.  Ele implementa diversas funções de processamento de imagens complementares aos do `ndimage`.  Alguma delas são funções de escalonamento das intensidades dos pixels, ou seja, alteram o histograma da imagem.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.ndimage as ndi
import PIL
```

```python
import skimage.exposure as exposure
import skimage.color as color

im = np.asarray(PIL.Image.open("girl.jpg"))
im1 = exposure.rescale_intensity(im, in_range=(75, 200))
plt.subplot(121)
fig = plt.imshow(im)
fig.axes.set_axis_off()
plt.subplot(122)
fig = plt.imshow(im1)
fig.axes.set_axis_off()
```

```python
plt.subplot(221)
im_gray = color.rgb2gray(im)
fig = plt.imshow(im_gray, cmap="gray")
fig.axes.set_axis_off()
plt.subplot(222)
hist = exposure.histogram(im_gray)[::-1]
_ = plt.plot(*hist)
plt.subplot(223)
im1_gray = color.rgb2gray(im1)
fig = plt.imshow(im1_gray, cmap="gray")
fig.axes.set_axis_off()
plt.subplot(224)
hist1 = exposure.histogram(im1_gray)[::-1]
hist1[1][:5] = hist1[1][-5:] = 0
_ = plt.plot(*hist1)
```

## Segmentação de imagens

Segmentação de uma imagem é o processo de separação de objetos de interesse do plano de fundo da imagem.  Em geral isto é feito com um esquema usando limiares, cuja versão mais simples pode ser feita com numpy.

```python
from pathlib import Path

with Path("iogurte.jpg") as imfile:
    im = np.asarray(PIL.Image.open(imfile))
plt.subplot(121)
fig = plt.imshow(im)
fig.axes.set_axis_off()
plt.subplot(122)
fig = plt.imshow(im[:,:,1], cmap="gray")
fig.axes.set_axis_off()
```

```python
im2 = np.where(im[:,:,1] > 100, 1, 0)
plt.imshow(im2, cmap="gray")
plt.show()
```

# Scikit Image - filtros não-lineares

## Processamento de imagens morfológico

Este tipo de processamento de imagens usa operações não-lineares relacionadas ao formato, ou morfologia, dos objetos da imagem.

O módulo `skimage` implementa diversas funções e estruturas de dados relativas a estas operações.

Em termos de estruturas de dados, o módulo de morfologia define vários arrays que podem ser usados como elementos estruturantes nas operações de filtros.  A maioria é bidimensional, como o `disk`, `diamond`, `rectangle`, `square` e `star`, mas alguns são tridimensionais, como `cube` e `octahedron`.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import skimage as ski
import skimage.morphology as morph
from skimage.morphology import disk
import skimage.filters.rank as rank
import skimage.filters as filters
import skimage.feature as feature


symmetric_octagon = lambda n: morph.octagon(n, n)
sizes = [1, 2, 3, 5, 10, 20]
elementos = [(morph.disk, "Disco"),
             (morph.star, "Estrela"),
             (morph.diamond, "Diamante"),
             (symmetric_octagon, "Octágono")]
for e, nome in elementos:
    for idx, n in enumerate(sizes):
        plt.subplot(231 + idx)
        plt.imshow(e(n), cmap="gray", interpolation="none")
        plt.title("{} tamanho {}".format(nome, n))
    plt.show()

for n in sizes:
    for idx, (e, nome) in enumerate(elementos):
        plt.subplot(221 + idx)
        plt.imshow(e(n), cmap="gray", interpolation="none")
        plt.title("{} de tamanho {}".format(nome, n))
    plt.show()
```

As operações básicas da morfologia são a dilação (expansão das áreas claras) e erosão (redução das áreas claras). A partir delas definimos outras operações compostas como o fechamento (`closing`), abertura (`opening`), `tophat` e `bottomhat`.

A operação de `dilation` associa a um pixel o máximo valor dentre todos os pixels da vizinhança definido pelo elemento estruturante. Assim, o `dilation` expande as áreas claras em detrimento das áreas escuras.

```python
base = np.asarray(PIL.Image.open("base.png"))
out = morph.dilation(base, disk(3))
plt.subplot(121)
fig = plt.imshow(base, cmap="gray")
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(122)
fig = plt.imshow(out, cmap="gray")
fig.axes.set_axis_off()
plt.title("Dilated")
plt.show()
```

A operação de `erosion` é o inverso da `dilation`; ela associa ao pixel o menor valor de intensidade na região definida pelo elemento estruturante.  Assim, esta operação tende a encolher as áreas claras e favorecer as áreas escuras.

```python
base = np.asarray(PIL.Image.open("base.png"))
out = morph.erosion(base, disk(3))
plt.subplot(121)
fig = plt.imshow(base, cmap="gray")
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(122)
fig = plt.imshow(out, cmap="gray")
fig.axes.set_axis_off()
plt.title("Eroded")
plt.show()
```

Ambas as operações podem ser aplicadas a imagens coloridas também

```python
garota = np.asarray(PIL.Image.open("girl.jpg"))
out1 = np.zeros_like(garota)
out2 = np.zeros_like(garota)
for n in range(3):
    out1[:,:,n] = morph.dilation(garota[:,:,n], disk(10))
    out2[:,:,n] = morph.erosion(garota[:,:,n], disk(10))
plt.subplot(131)
fig = plt.imshow(garota)
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(132)
fig = plt.imshow(out1)
fig.axes.set_axis_off()
plt.title("Dilated")
plt.subplot(133)
fig = plt.imshow(out2)
fig.axes.set_axis_off()
plt.title("Eroded")
plt.show()
```

Quando combinamos a operação de dilação com a de erosão, temos a operação denominada `closing`.

```python
base = np.asarray(PIL.Image.open("base.png"))
out = morph.closing(base, disk(10))
plt.subplot(121)
fig = plt.imshow(base, cmap="gray")
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(122)
fig = plt.imshow(out, cmap="gray")
fig.axes.set_axis_off()
plt.title("Closing")
plt.show()
```

A nomeclatura fica mais clara quando temos uma imagem com pequenas regiões escuras

```python
noise = np.asarray(PIL.Image.open("noise2.png"))
out = morph.closing(base, disk(3))
plt.subplot(121)
fig = plt.imshow(noise, cmap="gray")
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(122)
fig = plt.imshow(out, cmap="gray")
fig.axes.set_axis_off()
plt.title("Closing")
plt.show()
```

Quando fazemos o inverso,. i.e., combinamos a operação de erosão com a de dilação, temos a operação denominada `opening`.

Novamente, a nomeclatura fica mais clara quando temos uma imagem com pequenas regiões claras

```python
noise = np.asarray(PIL.Image.open("noise3.png"))
out = morph.opening(base, disk(3))
plt.subplot(121)
fig = plt.imshow(noise, cmap="gray")
fig.axes.set_axis_off()
plt.title("Original")
plt.subplot(122)
fig = plt.imshow(out, cmap="gray")
fig.axes.set_axis_off()
plt.title("Opening")
plt.show()
```
